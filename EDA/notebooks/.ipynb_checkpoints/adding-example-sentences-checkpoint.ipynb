```python
import pandas as pd 
import requests
import nltk
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
raw_df = pd.DataFrame(pd.read_csv("../../nytwit_v1.tsv", "\t"))
col_names = raw_df.columns
```

```python
raw_df.columns = raw_df.columns.str.replace(' ', '_')
raw_df
```

# Testing out the getText() function in BeautifulSoup

```python
url_list = [url for url in raw_df['URL'].values]
test_url = url_list[1]

html = urlopen(test_url).read()

soup = BeautifulSoup(html, 'html.parser')
textToParse = soup.getText()
textToParse
```

```python
clean_url_list = [url for url in url_list if url != 'not found']
len(clean_url_list) < len(url_list)
```

```python
list_of_sentences = []
list_of_indexes = []
for i, vals in enumerate(raw_df[['Word', 'URL']].values):
    try:
        
        if str(vals[-1]) != 'not found':    
            html = urlopen(str(vals[-1])).read()
            soup = BeautifulSoup(html, 'html.parser')
            textToParse = soup.getText()
            sentence_tokenize = nltk.sent_tokenize(textToParse)
            sentence_to_add = [sent for sent in sentence_tokenize if str(vals[0]) in sent]
            list_of_sentences.append(sentence_to_add)
            list_of_indexes.append(i)
            print(len(list_of_indexes), len(list_of_sentences))
        else:
            list_of_sentences.append("null")
            list_of_indexes.append(i)
            print(len(list_of_indexes), len(list_of_sentences))
            continue
    except:
        list_of_sentences.append("null")
        list_of_indexes.append(i)
        print(len(list_of_indexes), len(list_of_sentences))
        continue
```

```python
def zip_lists(list1, list2):
    return zip(list1, list2)
zip_lists(list_of_indexes, list_of_sentences)
```

```python
df = pd.DataFrame(zip_lists(list_of_indexes, list_of_sentences), columns = ["Indexes","Sentence Example"])
df.shape
```

```python
final_df = pd.concat([raw_df, df], axis=1)
final_df = final_df.drop(columns = ['Indexes'])
final_df.head()
```

```python
final_df['Sentence Example'] = [''.join(val) for val in final_df['Sentence Example']]
final_df
```

```python
cleaning_values = [val[1:] for val in df['Sentence Example'] if val[:1]==")"]


def get_messy_vals(df, colName):
    i = 0
    index_list = []
    for val in df[colName]:
        if val[:1] ==")":
            index_list.append(i)
            i+=1
        else:
            i+=1
            continue
    return index_list

for val in get_messy_vals(final_df, "Sentence Example"):
    final_df.iloc[val][-1] = final_df.iloc[val][-1][1:]
r, c = final_df.shape
print("Rows: " + str(r) + ", Columns: "+ str(c))
```

```python
final_df.to_csv('../../nytwit_v2', index=False, header=True)
```
